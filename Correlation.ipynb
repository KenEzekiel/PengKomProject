```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Dataframe declaration
url1 = 'https://raw.githubusercontent.com/KenEzekiel/PengKom-Tubes-2/main/AAPL%20Historical%20Data.csv?token=AVF4CUY3EGLLBJ2KHRPX52TBVGTHA'
df1 = pd.read_csv(url1)

url2 = 'https://raw.githubusercontent.com/KenEzekiel/PengKom-Tubes-2/main/GOOGL%20Historical%20Data.csv?token=AVF4CU6AU7CT5VZBTJK77S3BVGYEG'
df2 = pd.read_csv(url2)

url3 = 'https://raw.githubusercontent.com/KenEzekiel/PengKom-Tubes-2/main/NASDAQ%20Composite%20Historical%20Data%20(1).csv?token=AVF4CU4M7XR5B5NNYBEJSFDBVG57M'
df3 = pd.read_csv(url3)

# Data Cleansing
df1['Change %'] = df1['Change %'].str.rstrip('%').astype(float)
df1['Vol.'] = df1['Vol.'].str.rstrip('M').astype(float)

df2['Price'] = df2['Price'].str.replace(",","").astype(float)
df2['Open'] = df2['Open'].str.replace(",","").astype(float)
df2['High'] = df2['High'].str.replace(",","").astype(float)
df2['Low'] = df2['Low'].str.replace(",","").astype(float)
df2['Change %'] = df2['Change %'].str.rstrip('%').astype(float)
df2['Vol.'] = df2['Vol.'].str.rstrip('M').astype(float)

df3['Price'] = df3['Price'].str.replace(",","").astype(float)
df3['Open'] = df3['Open'].str.replace(",","").astype(float)
df3['High'] = df3['High'].str.replace(",","").astype(float)
df3['Low'] = df3['Low'].str.replace(",","").astype(float)
df3['Change %'] = df3['Change %'].str.rstrip('%').astype(float)
df3['Vol.'] = df3['Vol.'].str.rstrip('M').astype(float)

# Column declaration
column_1_d = df1["Date"]
column_2_d = df2["Date"]
column_3_d = df3["Date"]

column_1_p = df1["Price"]
column_2_p = df2["Price"]
column_3_p = df3["Price"]

column_1_c = df1["Change %"]
column_2_c = df2["Change %"]
column_3_c = df3["Change %"]

column_1_v = df1["Vol."]
column_2_v = df2["Vol."]
column_3_v = df3["Vol."]

# Correlation number
correlation_p12 = round(column_1_p.corr(column_2_p), 2)   # Price of AAPL and GOOGL
correlation_c12 = round(column_1_c.corr(column_2_c), 2)   # Change % of AAPL and GOOGL
correlation_v12 = round(column_1_v.corr(column_2_v), 2)   # Volume of AAPL and GOOGL

correlation_p13 = round(column_1_p.corr(column_3_p), 2)   # Price of AAPL and IXIC
correlation_c13 = round(column_1_c.corr(column_3_c), 2)   # Change % of AAPL and IXIC
correlation_v13 = round(column_1_v.corr(column_3_v), 2)   # Volume of AAPL and IXIC

correlation_p23 = round(column_2_p.corr(column_3_p), 2)   # Price of GOOGL and IXIC
correlation_c23 = round(column_2_c.corr(column_3_c), 2)   # Change % of GOOGL and IXIC
correlation_v23 = round(column_2_v.corr(column_3_v), 2)   # Volume of GOOGL and IXIC

# Printing the correlation for every combination
print("Price correlation for AAPL and GOOGL", correlation_p12)
print("Change % correlation for AAPL and GOOGL", correlation_c12)
print("Volume correlation for AAPL and GOOGL", correlation_v12)
print("")
print("Price correlation for AAPL and IXIC", correlation_p13)
print("Change % correlation for AAPL and IXIC", correlation_c13)
print("Volume correlation for AAPL and IXIC", correlation_v13)
print("")
print("Price correlation for GOOGL and IXIC", correlation_p23)
print("Change % correlation for GOOGL and IXIC", correlation_c23)
print("Volume correlation for GOOGL and IXIC", correlation_v23)
print("")

# Plotting graphs
df1.plot(kind='line', x='Date', y='Price', figsize=(15, 5), title="AAPL Price vs Date")
df2.plot(kind='line', x='Date', y='Price', color='red', figsize=(15, 5), title="GOOGL Price vs Date")
df3.plot(kind='line', x='Date', y='Price', color='green', figsize=(15, 5), title="IXIC Price vs Date")

df1.plot(kind='line', x='Date', y='Vol.', figsize=(15, 5), title="AAPL Volume vs Date")
df2.plot(kind='line', x='Date', y='Vol.', color='red', figsize=(15, 5), title="GOOGL Volume vs Date")
df3.plot(kind='line', x='Date', y='Vol.', color='green', figsize=(15, 5), title="IXIC Volume vs Date")

plt.show()

print(df1)
print(df2)
print(df3)
```
